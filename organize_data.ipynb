```python
import pyreadr
import pandas as pd
from joblib import load, dump 
```

```python
cat = ['Cho', 'George', 'Hagrid', 'Harry', 'Hermione', 'JellyB', 'Jube', 'Merry', 'MrsNorris', 'Neville', 'Nimbus', 'Scabbers']

dataset = pd.DataFrame(columns=['Timestamp', 'X_Mean', 'X_Min', 'X_Max', 'X_Sum', 'X_sd', 'X_Skew',
        'X_Kurt', 'Y_Mean', 'Y_Min', 'Y_Max', 'Y_Sum', 'Y_sd', 'Y_Skew',
        'Y_Kurt', 'Z_Mean', 'Z_Min', 'Z_Max', 'Z_Sum', 'Z_sd', 'Z_Skew',
        'Z_Kurt', 'VM_Mean', 'VM_Min', 'VM_Max', 'VM_Sum', 'VM_sd', 'VM_Skew',
        'VM_Kurt', 'ODBA15_Mean', 'ODBA30_Mean', 'ODBA45_Mean', 'ODBA60_Mean',
        'ODBA90_Mean', 'Cor_XY', 'Cor_XZ', 'Cor_YZ', 'Class'])

for name in cat:
    label = load(f"./cat_data/{name}_label").reset_index(drop=True)
    attri = load(f"./cat_data/{name}_attri")
    #Drop data
    label.drop(["Timestamp", "Pen", "Cat_id"], axis=1, inplace=True)
    attri.drop(["Cat_id", "Position"], axis=1, inplace=True)
    #process and connect
    label['Class'] = label.idxmax(axis=1)
    combined = pd.concat([attri, label['Class']], axis=1)
    dataset = pd.concat([dataset, combined], axis=0)
dataset.reset_index(drop=True, inplace=True)
```

```python
dataset = dataset[~dataset['Class'].isin(['Other_Social.Allogrooming', 'Other_Start', 'Other_Outofsight', 'Other_Other', 'Other_ActigraphOff'])]
display(dataset["Class"].unique())
```

```python
dataset.loc[dataset['Class'].isin(['Active_Playfight.Fighting', 'Active_Rubbing', 'Active_Playfight.Playing', 'Maintenance_Shake.Body', 'Active_Jumping.Vertical', 'Maintenance_Shake.Head', 'Active_Jumping.Horizontal', 'Active_Climbing']), 'Class'] = "active"
dataset.loc[dataset['Class'].isin(['Inactive_Standing.Stationary', 'Inactive_Standing.Up']), 'Class'] = "standing"
dataset.loc[dataset['Class'].isin(['Inactive_Sitting.Stationary', 'Inactive_Sitting.Down', 'Inactive_Sitting.Up']), 'Class'] = "sitting"
dataset.loc[dataset['Class'].isin(['Inactive_Lying.Crouch', 'Inactive_Lying.Down', 'Inactive_Lying.Resting']), 'Class'] = "lying"
dataset.loc[dataset['Class'].isin(['Active_Walking', 'Active_Trotting']), 'Class'] = "walk"
dataset.loc[dataset['Class'].isin(['Maintenance_Nutrition.Eating']), 'Class'] = "eat"
dataset.loc[dataset['Class'].isin(['Maintenance_Grooming']), 'Class'] = "groom"
dataset.loc[dataset['Class'].isin(['Maintenance_Scratching']), 'Class'] = "scratching"
dataset.loc[dataset['Class'].isin(['Maintenance_Littering.Digging', 'Maintenance_Littering.None', 'Maintenance_Littering.Urinating', 'Maintenance_Littering.Defecating']), 'Class'] = "other"
display(dataset["Class"].unique())
dataset.to_csv("./data/labeled_dataset1.csv", index=False)
```
